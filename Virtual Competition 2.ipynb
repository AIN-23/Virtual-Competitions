# Pre-processing on the training data

```python
## To find out if there are any missing or null values

train_data.isnull().sum()
```

```python
## To check if there is anu duplicate rows

len(train_data[train_data.duplicated()])
```

```python
train_data['Dependents'].value_counts()
```

```python
## To check the distribution of a particular column

# Create a histogram using Seaborn
sns.histplot(data=train_data, x='ApplicantIncome', bins=30, kde=True, color='red')

# Add labels and title
plt.xlabel('Applicant Income')
plt.ylabel('Frequency')
plt.title('Histogram to find the distribution')

# Show the plot
plt.show()
```

```python
## This is a right-skewed distribution.
```

```python
# Replace '3+' with the numeric value 3
# This assumes that having 3 or more dependents is a distinct category from having exactly 3 dependents.

train_data['Dependents'] = train_data['Dependents'].replace('3+', 3).astype(float)
train_data.head(10)
```

```python
## Filling the missing values in numeric columns.

# Fill the missing values with the mean
train_data['LoanAmount'].fillna(train_data['LoanAmount'].mean(), inplace=True)
train_data['Loan_Amount_Term'].fillna(train_data['Loan_Amount_Term'].mean(), inplace=True)

# Fill the missing values with the median
train_data['Credit_History'].fillna(train_data['Credit_History'].median(), inplace=True)
train_data['Dependents'].fillna(train_data['Dependents'].median(), inplace=True)

## Filling the missing values in categorical columns

# Fill the missing values with mode
train_data['Gender'].fillna(train_data['Gender'].mode()[0], inplace=True)
train_data['Married'].fillna(train_data['Married'].mode()[0], inplace=True)
train_data['Self_Employed'].fillna(train_data['Self_Employed'].mode()[0], inplace=True)
```

```python
## There are no null values now

train_data.isnull().sum()
```

```python
## Checking for outlier now using box plot

# Store the column names for future use
outlier_feature = ('Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History')

# Plot the box plot of feature with outliers
plt.figure(figsize=(20,10))
sns.boxplot(train_data[list(outlier_feature)],palette='plasma')
plt.title("Features with Outliers")
```

```python
## To remove the outliers

# Extract numeric columns
numeric_columns = train_data.select_dtypes(include=['float64', 'int64']).columns

# Function to replace outliers in each column with the mean of that column
def replace_outliers_with_mean(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Replace outliers with the mean of the column
    column[(column < lower_bound) | (column > upper_bound)] = column.mean()
    return column

# Apply the function to each numeric column
train_data[numeric_columns] = train_data[numeric_columns].apply(replace_outliers_with_mean, axis=0)

# Display box plots to check for outliers
plt.figure(figsize=(20, 10))
sns.boxplot(data=train_data[numeric_columns])
plt.title('Box Plot of Numeric Columns after Outlier Replacement')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

```python
train_data.head(3)
```

```python
## Dropping the first column since it will be of no use in later stages.
# It has all unique IDs

train_data.drop('Loan_ID', axis=1, inplace=True)
train_data.head(3)
```

# Encoding the Data

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical columns
categorical_columns = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']

# Create a LabelEncoder instance
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
for column in categorical_columns:
    if column in train_data.columns:
        train_data[column] = label_encoder.fit_transform(train_data[column])

# Display the DataFrame after label encoding
train_data.head(10)
```

# Scaling the dataset

```python
from sklearn.preprocessing import Normalizer

# Get the column name for the future use
numeric_data = train_data.iloc[:,:-1]
categoric_data=train_data['Loan_Status']

# Initiatiate and fit scaler
train_scaler = Normalizer()
train_scaled = train_scaler.fit_transform(numeric_data)

# Embed the scaled data into the dataset
train_scaled_df = pd.DataFrame(train_scaled,columns=numeric_data.columns)

train_scaled_data=pd.concat([train_scaled_df,categoric_data],axis=1)
train_scaled_data.head()
```
